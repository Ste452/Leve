```python
import pandas as pd

#Abre o arquivo de Groupby como DataFrame
vendas_DF = pd.read_excel("Groupby.xlsx")

#mean = Média
#groupby = Agrupando
mediaVendedor = vendas_DF.groupby(["Vendedor"]).mean()

#Imprimindo os dados
display(mediaVendedor)
```

```python
import pandas as pd

#Abre o arquivo de Groupby como DataFrame
vendas_DF = pd.read_excel("Groupby.xlsx")

#sum = Soma
#groupby = Agrupando
somaVendedor = vendas_DF.groupby(["Vendedor"]).sum()

#Imprimindo os dados
display(somaVendedor)
```

```python
import pandas as pd

#Abre o arquivo de Groupby como DataFrame
vendas_DF = pd.read_excel("Groupby.xlsx")

#sum = Soma
#groupby = Agrupando
#dropna=False = Não deleto os valores em branco
deixandoValoresEmBranco = vendas_DF.groupby(["Vendedor"], dropna=False).sum()

#Imprimindo os dados
display(deixandoValoresEmBranco)
```
